```python
from bs4 import BeautifulSoup

# HTML content as a string

table = './Table.txt'
html = open(table, 'r')
html_content = html.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all <a> tags in the HTML
a_tags = soup.find_all('a')
#create a list to store all of the a_tags
downloadable_links = []

# Extract the href attribute value from each <a> tag
for a_tag in a_tags:
    # print(type(a_tag))
    href = a_tag['href']
    downloadable_links.append(href)

#view all of the links in the list
# for link in downloadable_links:
#     print(link)

#only select sales reports
downloadable_links = downloadable_links[4:24]


```

```python
import requests
import os

# Define the directory to save all of the downloaded files
downloadeddir = './Downloads/'


# test sample
downloadable_linksSample = downloadable_links[0:2]
print(downloadable_linksSample)

for file_link in downloadable_linksSample:
    file_url = file_link
    file_name = file_link.split('/')[-1]
    file_name = file_name[8:]
    file_path = os.path.join(downloadeddir, file_name)

    # Send a GET request to download the file
    file_response = requests.get(file_url)
    response_code = file_response.status_code
    print(response_code)

    if file_response.status_code ==200:

        # Open a file in binary write mode
        with open(file_path, 'wb') as file:
            # Write the response content to the file
            file.write(file_response.content)
        print(f"Content saved to {file_name}")

    else:
        # Handle request errors
        print("Request failed with status code:", file_response.status_code)
```
