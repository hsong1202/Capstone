```python
import pandas as pd
import os

xlsx_folder = '../Capstone/Washoe County Sales report - Copy'
csv_folder = '../Capstone/CSV folder'


for filename in os.listdir(xlsx_folder):
    if filename.endswith('.xlsx'):
        # Construct the file paths
        xlsx_path = os.path.join(xlsx_folder, filename)
        csv_path = os.path.join(csv_folder, os.path.splitext(filename)[0] + '.csv')
        
        # Read the XLSX file and convert it to CSV
        df = pd.read_excel(xlsx_path)
        df.to_csv(csv_path, index=False)

```
